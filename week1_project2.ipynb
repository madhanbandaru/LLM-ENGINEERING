```python
from openai import OpenAI
from IPython.display import display, Markdown,update_display
from scraper import get_allLinks, get_web_content


```

```python
def get_messages(url):
    webpage_content=get_web_content(url)
    messages = [
        {
            "role": "system","content":"""You are a business data analyst. 
            Your task is to summarize the factual business information found in the provided text. 
            Focus on services, pricing, and contact info. 
            Do not reproduce copyrighted text; instead, provide a bulleted business analysis."""
        },
        {
            "role": "user", 
            "content": f"Analyze the following business data and create a summary brochure:\n\n{webpage_content}"
        }
    ]
    return messages
```

```python
ai=OpenAI(base_url='http://localhost:11434/v1',api_key='ollama')
for i in ai.models.list():
    print(i.id)
```

```python
def get_brochure(url):
    try:
        stream=ai.chat.completions.create(model=input(),messages=get_messages(url),stream=True)
        dis=display(Markdown("typing...."),display_id=True)
        responce=""
        for updatingdis in stream:
            text=updatingdis.choices[0].delta.content or ''
            if text:
                responce+=text
                update_display(Markdown(responce),
                               display_id=dis.display_id)
    except Exception as e:
        print(e)
    
```

```python
get_brochure(input())
```
