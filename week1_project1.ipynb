```python
import os
from openai import OpenAI
from IPython.display import display,Markdown,update_display

```

if you are seeing this project in git repo kindly ignore this below code and enter you api_key next after code note 

```python
from dotenv import load_dotenv
load_dotenv(override=True)
api_key=os.getenv('api_key')
if not api_key:
    api_key=input('enter your api key')
```

```python
#api_key=''
```

```python
try:
    ai=OpenAI(api_key=api_key,base_url=' ' )#enter your url
    for i in ai.models.list():
        print(i.id)
except Exception as e:
    print(e)
```

copy one model from the list and run next note and paste it in the input 

Note :- you'll get a popup there you are going to write the input like model name and your qns

```python
try:
    model=input("enter the model you want to use") 
    messages=[{"role":"system","content":"you are a good assistant and slove the user problem with showing realtime exmaples"},{"role":"user","content":input("enter your problem or a question for the AI answer")}]
    stream=ai.chat.completions.create(messages=messages,model=model,stream=True)
    responce=""
    display_handle=display(Markdown("thinking..."),display_id=True)
    for chuck in stream:
        responce+= chuck.choices[0].delta.content or ''
        update_display(Markdown(responce),display_id=display_handle.display_id)
except Exception as e:
    print(e)
```
